```python
df = pd.read_csv('crimeincidentreports.csv')
df
```

```python
ids = list(df)
ids
```

```python
tcd = [[] for _ in range(17)]
tcd
```

```python
for k in range(17):
    df2 = set(df[ids[k]])
    tcd[k] = {dic:i for i,dic in enumerate(df2)}
```

```python
for i in range(17):
    df[ids[i]] = df[ids[i]].map(tcd[i])
```
